```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
basics.head()
```

```python
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
ratings.head()
```

```python
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
akas.head()
```

# Data Cleaning: Title Basics

```python
basics.head()
```

```python
basics_df = basics.copy()
```

```python
basics_df.head()
```

```python
basics_df.replace({'\\N':np.nan}, inplace = True)
```

```python
basics_df.head()
```

```python
basics_df = basics_df.dropna(subset = ['runtimeMinutes'])
```

```python
basics_df = basics_df.dropna(subset = ['genres'])
```

```python
filter_movie = basics_df['titleType'] == 'movie'
```

```python
basics_df_movie = basics_df.loc[filter_movie]
basics_df_movie.head()
```

```python
basics_df_movie.info()
```

```python
basics_df_movie['startYear'] = pd.to_numeric(basics_df_movie['startYear'])
```

```python
basics_df_movie.info()
```

```python
filter_year = basics_df_movie['startYear'].between(2000, 2022, inclusive = True)
```

```python
basics_df_movie_year = basics_df_movie.loc[filter_year]
basics_df_movie_year.head()
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics_df_movie_year['genres'].str.contains('documentary',case=False)
basics_df_movie_year_nodoc = basics_df_movie_year[~is_documentary]
```

```python
basics_df_movie_year_nodoc.head()
```

```python
basics_final = basics_df_movie_year_nodoc.copy()
```

# Data Cleaning: AKAs

```python
akas.info()
```

```python
akas.head()
```

```python
akas.replace({'\\N':np.nan}, inplace = True)
```

```python
filter_region = akas['region'] == 'US'
```

```python
akas_df = akas.loc[filter_region]
akas_df.head()
```

# Data Cleaning: Ratings

```python
ratings.replace({'\\N':np.nan}, inplace = True)
```

```python
ratings.info()
```

```python
ratings
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers_1 =ratings['tconst'].isin(akas['titleId'])
keepers_1
```

```python
ratings = ratings[keepers_1]
ratings

```
